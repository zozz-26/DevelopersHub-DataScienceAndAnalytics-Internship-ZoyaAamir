### DATASET DESCRIPTION:

Iris dataset contains 150 rows and 5 columns — sepal length, sepal width, petal length, petal width and species. It contains data of 3 flower species — setosa, versicolor, virginica.


### CONCLUSION:
Iris dataset contains data of 3 flower species. Through EDA, we analyzed dataset structure, statistics, and created 3 visualizations. Scatter plot
showed positive correlation between sepal and petal length. Setosa has
smallest petal size with least variation, virginica has largest. Box plot
revealed outliers in setosa and versicolor species.

```python
# importing Libraries
import numpy as np                      # for mathematical calculations
import pandas as pd                     # for handle data in tables, spredsheets or other formats
import seaborn as sns                   # for creating graphs and charts
import matplotlib.pyplot as plt         # base library for creating plots, seaborn uses it underneath
```

```python
df = sns.load_dataset('iris')       # loading iris dataset
```

```python
df.head()    # displays first 5 rows
```

```python
df.shape      #displays total number of rows and columns in a dataset
```

```python
df.columns      # displays all column names of dataset
```

```python
df.describe()       # displays statistics of the dataset (count values, average value of every column,variation, minmum value,maximum value, data distibution)
```

```python
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')     # scatter plot to analyze relationship between sepal and petal length by species
plt.title('Sepal Length Vs Petal Length')                                       # title of the graph
plt.show()                                                                      # Display the graph
```

```python
sns.histplot(data=df, x='sepal_length')                           # histogram to analyze distribution of sepal length
plt.title('Sepal Length Distribution')                            # title of the graph
plt.show()                                                        # displays the graph
```

```python
sns.boxplot(data=df, x='species', y='petal_length')             # boxplot to analyze distribution of petal length by species
plt.title('Petal Length Distribution by Species')                # title of the graph
plt.show()                                                       # displays the graph
```
